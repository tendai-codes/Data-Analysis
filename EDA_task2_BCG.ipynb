#Descriptive statisistics

```python
# pie chart showing Churn Rate
plt.figure(figsize=(10, 8))
client_df['churn'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Churn Distribution')
plt.legend(['No Churn', 'Churn'])
plt.show()
```

```python
# prompt: create stacked bar chart for churn rate by channel sales and include percentages

# Group the data by 'channel_sales' and 'churn' and count the occurrences
churn_by_channel = client_df.groupby(['channel_sales', 'churn'])['id'].count().unstack()

# Calculate the percentages
churn_by_channel_pct = churn_by_channel.div(churn_by_channel.sum(axis=1), axis=0) * 100

# Plot the stacked bar chart
ax = churn_by_channel_pct.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Churn Rate by Channel Sales')
plt.xlabel('Channel Sales')
plt.ylabel('Percentage')
plt.legend(title='Churn', loc='upper right')

# Add percentage labels to the bars
for p in ax.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy()
    ax.text(x + width / 2, y + height / 2, '{:.1f}%'.format(height), ha='center', va='center')

plt.show()
```

```python
# Group by 'channel_sales' and calculate churn rate
churn_by_channel = client_df.groupby('channel_sales')['churn'].sum()

# Pie chart
plt.figure(figsize=(10, 10))
plt.pie(churn_by_channel, labels=churn_by_channel.index, autopct='%1.1f%%')
plt.title('Churn Rate by Channel Sales')
plt.legend(loc='lower left', bbox_to_anchor=(-0.3, 0.15))
plt.show()
```

```python
def plot_stacked_bars(column, title):
    # Group by 'churn' and calculate the sum of the specified column
    grouped = temp_df.groupby('churn')[column].sum()

    # Calculate percentages
    total = grouped.sum()
    percentages = grouped / total * 100

    # Create stacked bar chart
    plt.bar(['No Churn', 'Churn'], percentages, color=['blue', 'orange'])
    plt.title(title)
    plt.ylabel('Percentage')

    # Label percentages
    for i, v in enumerate(percentages):
        plt.text(i, v/2, f"{v:.1f}%", ha='center', va='center', color='white')


```

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 8))

# Plot the first stacked bar chart
plot_stacked_bars('cons_12m', '% of Electricity consuming customers that churned')
plt.sca(axes[0]) # Set the current axes to the first subplot

# Plot the second stacked bar chart
plot_stacked_bars('cons_last_month', '% of Electricity consuming customers that churned')
plt.sca(axes[1]) # Set the current axes to the second subplot

plt.tight_layout()
plt.show()
```

```python
# Function for heat map visualisation to correlation of  integer variables that affect ‘churn’

def corr_heatmap(df, target_variable):
  # Select integer columns and the target variable
  int_df = df.select_dtypes(include=['int'])
  int_df[target_variable] = df[target_variable]

  # Calculate correlations
  corr = int_df.corr()

  # Create a mask to hide the upper triangle of the heatmap
  mask = np.triu(np.ones_like(corr, dtype=bool))

  # Set up the matplotlib figure
  plt.figure(figsize=(12, 10))

  # Generate a custom diverging colormap
  cmap = sns.diverging_palette(230, 20, as_cmap=True)

  # Draw the heatmap with the mask and correct aspect ratio
  sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
              square=True, linewidths=.5, cbar_kws={"shrink": .5}, annot=True)

  plt.title(f'Integer Correlation Heatmap (with {target_variable})')
  plt.show()


```

```python
corr_heatmap(client_df, 'churn')
```
